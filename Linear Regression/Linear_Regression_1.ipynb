Create a dataset of size 30x2, where the first column represents the number of hours
of sunshine and the second column represents the number of ice creams sold.

i. Write a Python program to display the first five rows. Print the total number of
rows and columns in the dataset, and the range of each featureâ€™s values.

ii. Write a Python program to randomly split the dataset into training and testing
sets using the following ratios: 70-30,80-20,90-10.

iii. Apply Simple Linear Regression to predict the number of ice creams sold based
on the number of hours of sunshine. Report which training and testing split
yields the best results.Validate your model using the Mean Squared Error (MSE).

```python
hours_of_sunshine = [7, 8, 7, 8, 2, 9, 1, 3, 2, 4, 6, 4, 1, 6, 8, 1, 9, 3, 2, 5, 12, 2, 3, 11, 11, 2, 3, 1, 8, 6]
icecreams_sold = [20, 30, 18, 25, 7, 35, 3, 11, 10, 14, 20, 11, 3, 19, 35, 4, 31, 11, 8, 24, 40, 4, 10, 39, 42, 5, 12, 2, 29, 25]
```

```python
import pandas as pd
data = {
    'Hours of Sunshine': hours_of_sunshine,
    'Ice Creams Sold': icecreams_sold
}

df = pd.DataFrame(data)
```

```python
# i
import numpy as np
print("First five rows: ")
print(df[['Hours of Sunshine', 'Ice Creams Sold']][0: 5])

print("Total rows: ", df.shape[0])
print("Total columns: ", df.shape[1])

max_col1 = np.max(np.array(df['Hours of Sunshine']))
min_col1 = np.min(np.array(df['Hours of Sunshine']))
range_col1 = max_col1 - min_col1

max_col2 = np.max(np.array(df['Ice Creams Sold']))
min_col2 = np.min(np.array(df['Ice Creams Sold']))
range_col2 = max_col2 - min_col2
print("Range of feature 1: ", range_col1)
print("Range of feature 2: ", range_col2)
```

```python
#ii
import random 
#70-30
random_arr_70_30 = random.sample(range(30),21 )
X = list(df['Hours of Sunshine'])
y = list(df['Ice Creams Sold'])
X_train_70 = []
X_test_30 =[]
y_train_70 =[]
y_test_30 =[]
for i in range(30):
    if i in random_arr_70_30:
        X_train_70.append(X[i])
        y_train_70.append(y[i])
    else:
        X_test_30.append(X[i])
        y_test_30.append(y[i])
        
#80 -20        
random_arr_80_20 = random.sample(range(30),24 )
X_train_80 = []
X_test_20 =[]
y_train_80 =[]
y_test_20 =[]
for i in range(30):
    if i in random_arr_80_20:
        X_train_80.append(X[i])
        y_train_80.append(y[i])
    else:
        X_test_20.append(X[i])
        y_test_20.append(y[i])
        
# 90 - 10
random_arr_90_10 = random.sample(range(30),27 )
X_train_90 = []
X_test_10 =[]
y_train_90 =[]
y_test_10 =[]
for i in range(30):
    if i in random_arr_90_10:
        X_train_90.append(X[i])
        y_train_90.append(y[i])
    else:
        X_test_10.append(X[i])
        y_test_10.append(y[i])

print(" 70 - 30 ")
print("X train: ", X_train_70)
print("X test", X_test_30)
print("Y train: ", y_train_70)
print("Y test: ", y_test_30)
print()
print(" 80 - 20 ")
print(" X train: ", X_train_80)
print("X test", X_test_20)
print("Y train: ", y_train_80)
print("Y test: ", y_test_20)
print()
print(" 90 - 10 ")
print(" X train: ", X_train_90)
print("X test", X_test_10)
print("Y train: ", y_train_90)
print("Y test: ", y_test_10)
```

```python
# 70-30
n = len(X_train_70)
sigma_x = 0
for i in X_train_70:
    sigma_x += i
    
sigma_y = 0
for i in y_train_70:
    sigma_y += i
    
sigma_x_y = 0
for i in range(n):
    sigma_x_y +=( X_train_70[i] * y_train_70[i])

sigma_x_2 = 0
for i in X_train_70:
    sigma_x_2 += i**2

numerator_c1 = sigma_x_y - ((sigma_y*sigma_x)/n)
denominator_c1 = sigma_x_2 - ((sigma_x**2)/n)
c1 = numerator_c1/denominator_c1
print("C1: ",c1)

sigma_c1_x = 0
for i in X_train_70:
    sigma_c1_x += (i*c1)

numerator_c0 = sigma_y - sigma_c1_x
denominator_c0 = n
c0 = numerator_c0/denominator_c0
print("C0: ",c0)

# Linear Regression
y_predict_30 =[]
for i in X_test_30:
    y_predict_30.append(c0 + c1*i)
    
print("Y Predicted for 70 - 30: ", y_predict_30)
print("Y Actual for 70 - 30: ", y_test_30)

squared_error_70_30 = 0
for i in range(len(y_test_30)):
    squared_error_70_30 += ((y_test_30[i] - y_predict_30[i])**2)

mean_squared_error_70_30 = squared_error_70_30/len(y_test_30)
print("Mean squared error for 70-30: ", mean_squared_error_70_30)
```
